## SEABORN CATEGORICAL DATA PLOTS

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = sns.load_dataset('tips')
df
```

### COUNT PLOT

```python
sns.countplot(x = df['sex'],hue=df['smoker'])
```

### BAR PLOT 

```python
sns.barplot(x  = df['sex'], y = df['total_bill'], color='brown')
# barplot me x-> qualitative, y-> quantitative. 
```

You can change the estimator object to your own function, that converts a vector to a scalar 
Mean (default) : np.mean
Median: np.median 
Sum: np.sum
Count: len(to count the number of ovservations) 
Standard Deviation: np.std
Minimum: np.min
Maximum: np.max 

```python
import  numpy as np
sns.barplot(x = df['sex'], y = df['tip'],estimator=np.sum, color='magenta')
```

### BOXPLOT 
USED TO SHOW DISTRIBUTION OF CATEGORICAL DATA.

```python
sns.boxplot(x= 'tip', y = 'day', data=df,  palette='rainbow')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data=df,)
```

### STRIP PLOT & SWARM PLOT

Strip plot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution. 

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don't overlap.  This gives a better representation of the distribution of values, although it does not scale as well to lare numbers of observations (both in terms  of the ability to show all the point and in terms of the computation needed to arrange them). 

```python
sns.stripplot(x = 'day', y = 'total_bill', data=df)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data=df)
```

#### combine swarm and violin plots 

```python
sns.violinplot(x = 'tip', y = 'day', data=df)
sns.swarmplot(x = 'tip', y = 'day', data=df)
sns.stripplot(x = 'tip', y = 'day', data=df)
```
