#### DISTRIBUTION PLOT
are created for quantitative variable. ex age, salary 
Lets discuss some plots that us to visualize the distribution of a data set. These plots are:
* joinplot
* pairplot
* rugplot
* distplot/histplot 

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

> you can load different dataset which is inbuilt in sns. it can be found in git hub

```python
df = sns.load_dataset('tips') 
df
```

```python
df['size'].unique() # size is categorical or quanlitative. 
```

```python
sns.histplot(df['total_bill'],bins = 20,kde = True) 
# kde = kernel density estimator
# kde is the blue line, it is used to understand the distribution of data-
#  - to see the skewness of the data.
# distplot and histplot are the same. 
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20, kde=True) 
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

### JOINT PLOT

> we are using only one attribute rn. But we want to compare total_bill and tip. we use jointplot().
> kisi do distributed data ya two quantitative variable ka distribution set krna hai tohh use joint plot. 

```python
sns.jointplot(x = 'total_bill', y = 'tip', data=df, kind='reg')  
```

### PAIR PLOT

```python
sns.pairplot(df, hue='size',palette='rainbow')
```

> total_bill, tip & size ka comparison. 

> hue is used to divide the data category wise. EX: MALE,FEMALE. 

### RUG PLOT

```python
sns.rugplot(df)
```
