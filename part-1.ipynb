```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,5,11) 
# linear spacing - 0 se 5 ke beech 11 elements
x
 
y = x**2 
y
```

```python
plt.plot(x,y) # it will plot x values first then y values.
# plot makes a line graph 
plt.title('hh')
plt.xlabel('chhh') # name of x axis use xlabel.
plt.ylabel('hhh') # name of y axis use ylabel.
```

#### Ek saath multiple plots banana chaah rahe hain use subplot.

```python
plt.subplot(2,2,1)  
plt.subplot(2,2,2)  
plt.subplot(2,2,3)  
plt.subplot(2,2,4)
```

```python
plt.subplot(2,2,1)  
plt.plot()
plt.subplot(2,2,2)  
plt.plot(x**2,x**2)
plt.subplot(2,2,3)  
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
plt.subplot(6,6,1)
plt.subplot(6,6,2)
plt.subplot(6,6,3)
plt.subplot(6,6,4)
plt.subplot(6,6,9)
plt.subplot(6,6,34)
```

#### object oriented way

> white background is known as canvas where we plot. 
we access this canvas and perform things. 

```python
fig = plt.figure(figsize=(8,8),dpi = 100) # used for changing the size of canvas
fig.add_axes([0.1,0.1,0.8,0.8]) 
# left se shift, bottom se shift, size 80 percent width, 80 percent height.
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1]) 
axis1.plot(x,y)  
axis2 = fig.add_axes([0.3,0.2,0.3,0.8]) 
axis2.plot(x,y) 
```
