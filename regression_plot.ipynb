## REGRESSION PLOT

Rgression Plot is used in machine learning so we will not fully discuss what, how and where it  is used but we will see how it is created what are the values we can put inside it. 

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.lmplot(x = 'total_bill',y = 'tip', data=tips)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip', data=tips, hue='sex')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip', data=tips, hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip', data=tips, hue='sex',palette='coolwarm',markers=['o','v'],
           scatter_kws={'s':50})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 
import cufflinks as cf
from plotly.offline import iplot
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import plotly.express as px
tips['total_bill'].iplot() 
```

```python
tips.groupby('day')['tip'].mean().iplot(kind='bar') 
```

```python
tips.iplot(kind='scatter',x='total_bill',y = 'tip')
```
